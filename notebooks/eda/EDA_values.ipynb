```python
import os
import pandas as pd
from tqdm import tqdm
from json import loads, load
```

Approach: </br>
- Create two dataframes with attributes as columns
- enter the actual values in one dataframe
- enter the datatype of values in another dataframe

```python
main_fol_path = "../data/external/from-vamsi"
component_list = os.listdir(main_fol_path)
meta_df = pd.DataFrame()
meta_type_df = pd.DataFrame()

for component in component_list:

    print(f"processing {component}")

    json_file_paths = os.listdir(f"{main_fol_path}/{component}")
    ### remove .png or any other file type
    json_file_paths = [x for x in json_file_paths if x.split('.')[1] == "json"]

    for path in tqdm(json_file_paths):
        json_file =  open(f"{main_fol_path}/{component}/{path}")

        ### 
        json_dict = load(json_file)

        json_type_dict = {key: type(value) for key,value in json_dict.items()}

        id = path.split('.')[0]
        json_dict["id"] = id
        json_type_dict["id"] = id

        # df1 = pd.DataFrame([data])
        meta_df = meta_df.append(pd.DataFrame([json_dict]), ignore_index = True)
        meta_type_df = meta_type_df.append(pd.DataFrame([json_type_dict]), ignore_index = True)
        
        json_file.close()
```

### Component wise exploratory analysis 

```python
def analyse_component(meta_df, component):
    df = meta_df[meta_df.componentType == component]
    df = df.dropna(axis=1, how='all')
    return df, df.columns
```

### Create an attribute availability table for each component

```python
lists_dict = {"switch_list":list(switch_list), "icon_list":list(icon_list), "checkbox_list":list(checkbox_list), 
              "button_list":list(button_list), "card_list":list(card_list)}
```

```python
df = pd.DataFrame(0, index = set(list(switch_list)+list(icon_list)+list(button_list)+list(card_list)+list(checkbox_list)), 
                  columns = ['switch_list', 'icon_list', 'button_list', 'card_list', 'checkbox_list'])
for col in df.columns:
    for idx in df.index:
        if idx in lists_dict[col]:
            df.loc[idx, col] = 1
        else:
            df.loc[idx, col] = 0

df.head(50)
```

```python
df.to_csv("../data/outputs/attribute_availability.csv")
```
